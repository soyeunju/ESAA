# 맷플롯립(Matplotlib)
## 맷플롯립의 pyplot 모듈의 이해
* matplotlib.pyplot
* pyplot은 plot()함수를 통해 선 그래프를 그릴 수 있음. 입력 인자 (x좌표에 해당하는 데이터 값,y좌표에 해당하는 여러 데이터 값)
* title()함수는 입력값으로 타이틀 문자열을 받아서 차트 타이틀을 설정할 수 있음
* show()함수를 호출하여 축을 포함한 선 그래프를 화면에 나타낼 수 있음

```python
# pyplot 모듈을 (관용적으로) plt라는 이름으로 import,
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[2,4,6]) #plot()을 통해 선 그래프를 그릴 수 있음
# 선그래프 타이틀을 Hello plot으로 설정
plt.title('Hello plot')
# 선 그래프를 출력
plt.show()
```

## pyplot의 두 가지 중요 요소- Figure와 Axes 이해
* 먼저 Figure 객체는 그림을 그리기 위한 캔버스의 역할.그림판의 크기를 조절한다든가, 플롯을 최종적으로 화면에 나타내는 역할을 수행.그림을 그리는 실제 작업에는 크게 연관되어 있지 않음.
* Axes는 실제 그림을 그리는 메서드들을 가짐. 그래서 대부분의 시각화를 수행하는 메서드들은 Axes에서 호출. 또한 X축, Y축, 타이틀, 범례 등의 속성 을 설정하는 데도 이용.

## Figure와 Axis의 활용
* 전체 그림판의 크기를 조절하는 데 사용. plt.figure(figsize=(10, 4))와같이 plt
모듈의 figure() 함수를 호출하면 Figure 객체가 반환됨. 이때 인자로 figsize=(가로크기, 세로크기) 입력하면 해당 크기를 가지는 Figure 객체를 반환.

```python
# plt.figureO는 주로 figure의 크기를 조절하는 데 사용됨.
# 아래는 figure 크기가 가로 10, 세로 4인 Figure 객체를 설정하고 반환함
plt.figure(figsize=(10,4))

plt.plot([1,2,3],[4,6,8])
plt.title('Hello plot')

plt.show()
```

## 여러 개의 plot을 가지는 subplot들을 생성하기
* plt.subplots() 함수는 주로 여러 개의 subplot을 생성하는 데 활용함. 즉 하나의 Figure상에서 여러 개의 그래프를 그릴 수 있게 만들어 주는데, 이때 개별 subplot은 하나의 Axes를 가지게 됨. 즉 plt.subplots()은 여러 개의 그래프를 그릴 수 있게 하는 1개의 Figure와여러 개의 Axes를 생성하고 반환함.

* plt.subplots(nrows=1, ncols=2, figsize=(6, 3)) 수행 시 Figure와 두 개의 Axes 객체를 반환하는데, 이때 반환되는 Axes 객체는 (axl, ax2)와 같이 튜플 형태로 반환받을 수 있음. ax1가 첫번째 Axes 객체를, ax2가 두번째 Axes 객체를 가리킴.


```python
# 아래는 두 개의 subplot, 즉 두 개의 Axes를 생성하되 열 방향으로 두 개를 배치함.
# 또한 전체 크기가 가로 6, 세로 3인 Figure 생성.
fig, (ax1,ax2) =plt.subplots(nrows=1,ncols=2,figsize=(6,3))
```

## pyplot의 plot() 함수를 이용해 선 그래프 그리기
* pyplot의 plot() 함수는 선 그래프를 그릴 때 활용됨
* X 좌표 값, Y 좌표 값으로 파이썬 리스트, 넘파이 ndarray, 판다스(Pandas)의 DataFrame/Series 모두 적용 가능
* 다만 입력되는 X 좌 표 값과 Y 좌표 값의 개수는 모두 같은 크기를 가져야 함

```python
import numpy as np
import pandas as pd

x_value = [1,2,3,4]
y_value = [2,4,6,8]

#x_value = np.array([1, 2, 3, 4]) # 넘파이 array로 테스트하려면 주석을 해제하고 실행
#y_value = np.array([2, 4, 6, 8]) # 넘파이 array로 테스트하려면 주석을 해제하고 실행


# 입력값으로 파이썬 리스트, numpy array 가능. x 좌표값과 y좌표값은 모두 같은 크기를 가져야 함.
plt.plot(x_value,y_value)
plt.show()
```

* plot( ) 함수는 color 인자를 이용해 선의 색깔을 변경할 수 있음

```python
# color 인자를 green으로 설정하여 선의 색깔을 녹색으로 변경.
plt.plot(x_value,y_value,color='green')
```

```python
plt.plot(x_value,y_value,color='red',marker='o',
         linestyle='dashed',linewidth=2,markersize=12)
```

## 축 명칭 설정, 축의 눈금(틱)값 회전, 범례(legend) 설정하기

* 축의 명칭: plt.xlabel, plt.ylabel

```python
plt.plot(x_value,y_value,color='red',marker='o',
         linewidth=2,markersize=12,linestyle='dashed')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

* 눈금(틱) 회전: xticks(),yticks()
* 보통 눈금값을 회전시키는 이유는, 축의 값이 문
자열이고 많은 눈금값이 있을 경우 눈금 간에 값이 중복되면서 눈금값이 제대로 보이지 않기 때문

```python
# x 좌표 값을 0~99까지 100개로 설정.
x_values=np.arange(100)
# y 좌표 값은 개별 x좌표값을 2배로 해서 100개로 설정.
y_values=2*x_values

plt.plot(x_values,y_values,color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.xticks(rotation=90)
plt.show()
```

* X축의 눈금 값을 좀 더 세밀하게 나타내고 싶으면 xticks() 함수의 인자로 ticks 값을 나타내고 싶은 값을 설정하면 됨

```python
x_value=np.arange(100)
y_value=x_value*2

plt.plot(x_value,y_value,color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.xticks(rotation=90,ticks=np.arange(0,100,5)) # ticks 인자로 눈금 값을 더 세밀하게 나눌 수 있음

plt.show()
```

* 범례(legend):범례는 여러 개의 그래프/차트를 한 Axes 내에서 표시 할 때 이들을 구분할 때 편리하게 사용
* plt.plot() 내에 label 인자값을 설정해 해당 플롯에 명칭을 부여한 후 plt.legend()를 호출해 범례를 표시

```python
x_value=np.arange(100)
y_value01=2*x_value
y_value02=4*x_value

plt.plot(x_value,y_value01,color='green',label='green line')
plt.plot(x_value,y_value02,color='red',label='red line')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Hello plot')

plt.legend()
```

```python
x_value_01 = np.arange(10)
y_value_01 = 2*x_value_01

plt.plot(x_value_01,y_value_01,color='red',linestyle='dashed',
         marker='o',label='red line')

# plt.bar : bar chart. plt.plot: line graph
plt.bar(x_value_01,y_value_01,color='green',label='bar plot')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.title('Hello plot')
plt.show()
```

```python
# Axes 객체 메서드를 사용해 시각화를 수행
figure = plt.figure()
# Axes 객체를 추출하여 ax 변수에 할당.
ax = plt.axes()

# plt.plot( )은 ax.plot( )으로 plt.bar( )는 ax.bar( )로 변경.
ax.plot(x_value_01,y_value_01,color='red',
        marker='o',linestyle='dashed',label='red line')
ax.bar(x_value_01,y_value_01,color='green',label='green bar')

# plt.xlabel( )은 ax.set_xlabel( )로, plt.ylabel( )은 ax.set_ylabel( )로 변경
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

# ax.set_legend()가 아니라 ax.legend()임.
ax.legend()
# plt.title( )을 plt.set_title( )로 변경
ax.set_title('Hello plot')

plt.show()
```

## 여러 개의 subplots들을 이용해 개별 그래프들을 subplot별로 시각화하기

* 여러 개의 subplots들을 이용하면 여러 개의 그래프들을 한꺼번에 시각화할 수 있음. 이는 데이터 를 분석하는 측면에서 매우 유용함.

* 여러 개의 subplots들을 이용해서 시각화를 구현할 때는 유의해야 할 사항이 있음. 바로 개별 Axes 객체를 직접 이용해서 시각화를 구현해야 함. 이는 비단 선 그래프나 막대 그래프 를 그리는 것에 국한되지 않고, 타이틀이나, 축의 명칭 눈금값, 범례를 설정할 때도 개별 Axes에 직접 이들을 적용.

```python
# Axes 객체들을 튜플이 아닌 배열 형태로 받아서 시각화
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(6,3))

# 반환된 Axes 객체들을 가지는 ax 변수는 튜플이 아니라 넘파이 ndarray 형태로 가지며 각각 개별 원소인 ax[0]는 왼쪽에 생성된 Axes 객체를, ax[1]은 오른쪽에 생성된 Axes 객체를 가리킴
```

```python
print('ax type:',type(ax))
print('ax[0] type:',type(ax[0]))
print('ax[1] type:',type(ax[1]))
```

```python
import numpy as np
x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02


fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(6,3))


# 개별 Axes 객체들을 각각 이용하여 선 그래프와 막대그래프를 그림.
ax[0].plot(x_value_01, y_value_01, color='red',
           marker='o',linestyle='dashed',label='red line')

ax[1].bar(x_value_02, y_value_02, color='green',
           label='green bar')

# 개별 Axes 객체 각각에 X축명을 설정.
ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')

# 개별 Axes 객체 각각에 범례를 설정
ax[0].legend()
ax[1].legend()

# 개별 Axes 객체 각각에 타이틀을 설정
ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')

plt.show()
```

```python
# nrows도 2 이상이고, ncols도 2 이상이 되면 1차원 배열 인덱싱 방식이 아니라 2차원 배열 인덱싱 방식.
# plot.subplots()로 반환 된 Axes 객체를 접근해야 함.

import numpy as np
x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(6,6))

ax[0][0].plot(x_value_01,y_value_01,color='red',marker='o',
             linestyle='dashed',label='red line')
ax[0][1].bar(x_value_02,y_value_02,color='green',label='green bar')
ax[1][0].plot(x_value_01,y_value_01,color='green',marker='o',
             linestyle='dashed',label='green line')
ax[1][1].bar(x_value_02,y_value_02,color='red',label='red bar')

ax[0][0].set_xlabel('a[0][0] x axis')
ax[0][1].set_xlabel('a[0][1] x axis')
ax[1][0].set_xlabel('a[1][0] x axis')
ax[1][1].set_xlabel('a[1][1] x axis')

ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

plt.show()
```
