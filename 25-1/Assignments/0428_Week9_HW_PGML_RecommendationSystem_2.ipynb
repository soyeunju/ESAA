```python
from ast import literal_eval ## ÌååÏù¥Ïç¨ ast Î™®ÎìàÏùò literal_eval() Ìï®ÏàòÎ•º ÌÜµÌï¥ Ïù¥ Î¨∏ÏûêÏó¥ÏùÑ Î¨∏ÏûêÏó¥Ïù¥ ÏùòÎØ∏ÌïòÎäî list[ dictl, dict2 ] Í∞ùÏ≤¥Î°ú ÎßåÎì§ Ïàò ÏûàÏùå
movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['keywords'] = movies_df['keywords'].apply(literal_eval)
```

```python
# Í≤∞Í≥º ÌôïÏù∏
movies_df[['genres','keywords']][:1]
```

```python
# ÎîïÏÖîÎÑàÎ¶¨ÏóêÏÑú 'name' ÌÇ§Ïóê Ìï¥ÎãπÌïòÎäî Í∞íÎßå Ï∂îÏ∂ú
movies_df['genres'] = movies_df['genres'].apply(lambda x : [ y['name'] for y in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x : [ y['name'] for y in x])
movies_df[['genres', 'keywords']][:1 ]
```

### Ïû•Î•¥ ÏΩòÌÖêÏ∏† Ïú†ÏÇ¨ÎèÑ Ï∏°Ï†ï
genresÎ•º Î¨∏ÏûêÏó¥Î°ú Î≥ÄÍ≤ΩÌïú Îí§ Ïù¥Î•º CountVectorizerÎ°ú ÌîºÏ≤ò
Î≤°ÌÑ∞ÌôîÌïú ÌñâÎ†¨ Îç∞Ïù¥ÌÑ∞ Í∞íÏùÑ ÏΩîÏÇ¨Ïù∏ Ïú†ÏÇ¨ÎèÑÎ°ú ÎπÑÍµê

```python
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizerÎ•º Ï†ÅÏö©ÌïòÍ∏∞ ÏúÑÌï¥ Í≥µÎ∞±ÏúºÎ°ú word Îã®ÏúÑÍ∞Ä Íµ¨Î∂ÑÎêòÎäî Î¨∏ÏûêÏó¥Î°ú Î≥ÄÌôò
# join() Î©îÏÑúÎìú : Î¶¨Ïä§Ìä∏ ÎÇ¥Ïùò Í∞Å ÏöîÏÜåÎ•º ÌïòÎÇòÏùò Î¨∏ÏûêÏó¥Î°ú Í≤∞Ìï©Ìï®
movies_df['genres_literal'] = movies_df['genres'].apply(lambda x:(' ').join(x))
count_vect = CountVectorizer(min_df = 0.0, ngram_range =(1,2))
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
print(genre_mat.shape)
```

```python
movies_df['genres_literal'][:2]
```

```python
# ÏΩîÏÇ¨Ïù∏ Ïú†ÏÇ¨ÎèÑ Í≥ÑÏÇ∞
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat,genre_mat)
print(genre_sim.shape)
print(genre_sim[:2])
```

```python
# Î†àÏΩîÎìúÎ≥ÑÎ°ú Ïú†ÏÇ¨ÎèÑÍ∞íÏù¥ ÎÜíÏùÄ Î†àÏΩîÎìúÏùò ÏúÑÏπòÍ∞í(Ïù∏Îç±Ïä§) Ï†ÄÏû•
genre_sim_sorted_ind = genre_sim.argsort()[:,::-1]
print(genre_sim_sorted_ind[:1])
```

0Î≤à Î†àÏΩîÎìúÏùò Í≤ΩÏö∞ ÏûêÏã†Ïù∏ 0Î≤à Î†àÏΩîÎìú Î•º Ï†úÏô∏ÌïòÎ©¥ 3494Î≤à Î†àÏΩîÎìúÍ∞Ä Í∞ÄÏû• Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÍ≥†, Í∑∏Îã§ÏùåÏù¥ 813Î≤à Î†àÏΩîÎìúÏù¥Î©∞, Í∞ÄÏû• Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÇÆÏùÄ Î†àÏΩîÎìúÎäî 2401Î≤à Î†àÏΩîÎìúÎùºÎäî Îúª

### Ïû•Î•¥ ÏΩòÌÖêÏ∏† ÌïÑÌÑ∞ÎßÅÏùÑ Ïù¥Ïö©Ìïú ÏòÅÌôî Ï∂îÏ≤ú

Ïû•Î•¥ Ïú†ÏÇ¨ÎèÑÏóê Îî∞Îùº ÏòÅÌôîÎ•º Ï∂îÏ≤úÌïòÎäî `find_sim_movie()` Ìï®Ïàò ÏÉùÏÑ±

ÏûÖÎ†•Í∞í
* Í∏∞Î∞ò Îç∞Ïù¥ÌÑ∞ (movies_df)
* Î†àÏΩîÎìúÎ≥Ñ Ïû•Î•¥ ÏΩîÏÇ¨Ïù∏ Ïú†ÏÇ¨ÎèÑ Ïù∏Îç±Ïä§(genre_sim_sorted_ind)
* Í≥†Í∞ùÏù¥ ÏÑ†Ï†ïÌïú Ï∂îÏ≤ú Í∏∞Ï§ÄÏù¥ ÎêòÎäî ÏòÅÌôî Ï†úÎ™©
* Ï∂îÏ≤úÌï† ÏòÅÌôî Í±¥Ïàò

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    # Ïù∏ÏûêÎ°ú ÏûÖÎ†•Îêú movies_df ÏóêÏÑú 'title' ÏπºÎüºÏù¥ ÏûÖÎ†•Îêú title_name Í∞íÏù∏ DataFrame Ï∂îÏ∂ú
    title_movie = df[df['title']==title_name]

    # title_nameÏùÑ Í∞ÄÏßÑ DataFrameÏùò index Í∞ùÏ≤¥Î•º ndarrayÎ°ú Î∞òÌôòÌïòÍ≥†
    # sorted_ind Ïù∏ÏûêÎ°ú ÏûÖÎ†•Îêú genre_sim_sorted_ind Í∞ùÏ≤¥ÏóêÏÑú Ïú†ÏÇ¨ÎèÑ ÏàúÏúºÎ°ú top_nÍ∞úÏùò index Ï∂îÏ∂ú
    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]

    # Ï∂îÏ∂úÎêú top_n index Ï∂úÎ†•, top_n indexÎäî 2Ï∞®Ïõê Îç∞Ïù¥ÌÑ∞ÏûÑ
    # dataframe ÏóêÏÑú indexÎ°ú ÏÇ¨Ïö©ÌïòÍ∏∞ ÏúÑÌï¥ÏÑú 1Ï∞®Ïõê arrayÎ°ú Î≥ÄÍ≤Ω
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)

    return df.iloc[similar_indexes]
```

```python
similar_movies = find_sim_movie(movies_df,genre_sim_sorted_ind,'The Godfather',10)
similar_movies[['title','vote_average']]
```

Îçî ÎßéÏùÄ ÌõÑÎ≥¥Íµ∞ÏùÑ ÏÑ†Ï†ïÌïú Îí§Ïóê ÏòÅÌôîÏùò ÌèâÏ†êÏóê Îî∞Îùº ÌïÑÌÑ∞ÎßÅÌï¥ÏÑú ÏµúÏ¢Ö Ï∂îÏ≤úÌïòÎäî Î∞©ÏãùÏúºÎ°ú Î≥ÄÍ≤Ω

'vote_average'Ïùò Í≤ΩÏö∞ 0~10Ï†ê ÏÇ¨Ïù¥Ïùò Ï†êÏàòÎ°ú Í¥ÄÍ∞ùÎì§Ïù¥ Î∂ÄÏó¨Ìïú Ï†êÏàò. but, ÏÜåÏàòÏùò Í¥ÄÍ∞ùÏù¥ ÌäπÏ†ï ÏòÅÌôîÏóê ÎÜíÏùÄ ÌèâÏ†êÏùÑ Î∂ÄÏó¨Ìïú Í≤ΩÏö∞ ÏôúÍ≥°Îêú Ìï¥ÏÑù Í∞ÄÎä•

```python
movies_df[['title', 'vote_average', 'vote_count']].sort_values('vote_average', ascending=False)[:10]
```

Ïú†Î™ÖÌïú ÏòÅÌôî ÌèâÏ†ê ÏÇ¨Ïù¥Ìä∏Ïù∏ IMDBÏóêÏÑúÎäî ÌèâÍ∞Ä ÌöüÏàòÏóê ÎåÄÌïú Í∞ÄÏ§ëÏπòÍ∞Ä Î∂ÄÏó¨Îêú ÌèâÏ†ê(Weighted Rating) Î∞©ÏãùÏùÑ ÏÇ¨Ïö©
* Í∞ÄÏ§ë ÌèâÏ†ê(Weighted Rating) = $(v/(v+m))*R + (m/(v+m))*C$
  * $v$: Í∞úÎ≥Ñ ÏòÅÌôîÏóê ÌèâÏ†êÏùÑ Ìà¨ÌëúÌïú ÌöüÏàò (vote_count)
  * $m$: ÌèâÏ†êÏùÑ Î∂ÄÏó¨ÌïòÍ∏∞ ÏúÑÌïú ÏµúÏÜå Ìà¨ÌëúÌöüÏàò (m Í∞íÏùÑ ÎÜíÏù¥Î©¥ ÌèâÏ†ê Ìà¨Ìëú ÌöüÏàòÍ∞Ä ÎßéÏùÄ ÏòÅÌôîÏóê Îçî ÎßéÏùÄ Í∞ÄÏ§ë ÌèâÏ†êÏùÑ Î∂ÄÏó¨)
  * $R$: Í∞úÎ≥Ñ ÏòÅÌôîÏóê ÎåÄÌïú ÌèâÍ∑† ÌèâÏ†ê(vote_average)
  * $C$: Ï†ÑÏ≤¥ ÏòÅÌôîÏóê ÎåÄÌïú ÌèâÍ∑† ÌèâÏ†ê(mean(vote_average))

```python
C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6)
print('C:', round(C, 3), 'm:', round(m, 3))
```

```python
percentile=0.6
m = movies_df['vote_count'].quantile(percentile)
C = movies_df['vote_average'].mean()

def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']

    return ( (v/(v+m)) * R + (m/(m+v)) * C )

movies_df['weighted_vote'] = movies_df.apply(weighted_vote_average, axis=1)
```

```python
# Í∞ÄÏ§ëÌèâÏ†ê Í∏∞Ï§ÄÏúºÎ°ú ÏÉÅÏúÑ 10Í∞ú ÏòÅÌôî Ï∂îÏ∂ú
movies_df[['title','vote_average','weighted_vote','vote_count']].sort_values(by='weighted_vote', ascending=False)[:10]
```

Ïû•Î•¥ Ïú†ÏÇ¨ÏÑ±Ïù¥ ÎÜíÏùÄ ÏòÅÌôîÎ•º top_nÏùò 2Î∞∞ÏàòÎßåÌÅº ÌõÑÎ≥¥Íµ∞ÏúºÎ°ú ÏÑ†Ï†ïÌïú Îí§Ïóê weighted_vote ÏπºÎüº Í∞íÏù¥ ÎÜíÏùÄ ÏàúÏúºÎ°ú top_nÎßåÌÅº Ï∂îÏ∂úÌïòÎäî`Î∞©ÏãùÏúºÎ°ú find_sim_movie()` Ìï®Ïàò ÏàòÏ†ï

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie = df[df['title']==title_name]
    title_index = title_movie.index.values

    # top_nÏùò 2Î∞∞Ïóê Ìï¥ÎãπÌïòÎäî Ïû•Î•¥ Ïú†ÏÇ¨ÏÑ±Ïù¥ ÎÜíÏùÄ Ïù∏Îç±Ïä§ Ï∂îÏ∂ú
    similar_indexes = sorted_ind[title_index, :(top_n*2)]
    similar_indexes = similar_indexes.reshape(-1)
    # Í∏∞Ï§Ä ÏòÅÌôî Ïù∏Îç±Ïä§ Ï†úÏô∏
    similar_indexes = similar_indexes[similar_indexes != title_index]

    # top_nÏùò 2Î∞∞Ïóê Ìï¥ÎãπÌïòÎäî ÌõÑÎ≥¥Íµ∞ÏóêÏÑú weigthed_voteÍ∞Ä ÎÜíÏùÄ ÏàúÏúºÎ°ú top_nÎßåÌÅº Ï∂îÏ∂ú
    return df.iloc[similar_indexes].sort_values(by='weighted_vote', ascending=False)[:top_n]

similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title','vote_average','weighted_vote']]
```

ÌòëÏóÖ ÌïÑÌÑ∞ÎßÅÏùÄ Ïù¥ ratings.csv Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ÏôÄ Í∞ôÏù¥ ÏÇ¨Ïö©ÏûêÏôÄ ÏïÑÏù¥ÌÖú Í∞ÑÏùò ÌèâÏ†ê(ÎòêÎäî Îã§Î•∏ Ïú†ÌòïÏùò Ïï°ÏÖò) Ïóê Í∏∞Î∞òÌï¥ Ï∂îÏ≤úÌïòÎäî ÏãúÏä§ÌÖú.

Ïù¥Î•º Íµ¨ÌòÑÌïòÍ∏∞ ÏúÑÌï¥ Î®ºÏ†Ä Î°úÏö∞(Ìñâ) Î†àÎ≤® ÌòïÌÉúÏùò ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏Î•º Îã§Ïùå Í∑∏Î¶º

```python
ratings = ratings[['userId','movieId','rating']]
ratings_matrix = ratings.pivot_table('rating',index = 'userId',columns = 'movieId')
ratings_matrix.head(3)
```

```python
# title ÏπºÎüºÎ™ÖÏùÑ ÏñªÍ∏∞ ÏúÑÌï¥ moviesÏôÄ Ï°∞Ïù∏
rating_movies = pd.merge(ratings,movies, on = 'movieId')

# columns='title‚ÄôÎ°ú title ÏπºÎüºÏúºÎ°ú ÌîºÎ≤ó ÏàòÌñâ.
ratings_matrix = rating_movies.pivot_table('rating', index='userId', columns='title')

# NaN Í∞íÏùÑ Î™®Îëê 0ÏúºÎ°ú Ï±ÑÏö∞Í∏∞
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix.head(3)
```

### ÏòÅÌôî Í∞Ñ Ïú†ÏÇ¨ÎèÑ ÏÇ∞Ï∂ú

* `cosine_similarity()` Ìï®ÏàòÎäî ÌñâÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÏÑúÎ°ú Îã§Î•∏ ÌñâÏùÑ ÎπÑÍµêÌïòÏó¨ Ïú†ÏÇ¨ÎèÑÎ•º ÏÇ∞Ï∂ú
* ratings_matrixÎäî userIdÍ∞Ä Í∏∞Ï§ÄÏù∏ Ìñâ Î†àÎ≤® Îç∞Ïù¥ÌÑ∞Î°ú, Ïù¥ Îç∞Ïù¥ÌÑ∞Ïóê `cosine_similarity()`Î•º Ï†ÅÏö©ÌïòÎ©¥ ÏòÅÌôî Í∞ÑÏùò Ïú†ÏÇ¨ÎèÑÍ∞Ä ÏïÑÎãå ÏÇ¨Ïö©Ïûê Í∞ÑÏùò Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎßåÎì§Ïñ¥ÏßÄÍ≤å Îê®.
* ratings_matrixÏùò Ï†ÑÏπò ÌñâÎ†¨Î°ú ÏΩîÏÇ¨Ïù∏ Ïú†ÏÇ¨ÎèÑÎ•º Íµ¨Ìï¥Ïïº Ìï®

```python
ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T.head(3)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(ratings_matrix_T,ratings_matrix_T)

# cosine_similarity()Î°ú Î∞òÌôòÎêú ÎÑòÌååÏù¥ ÌñâÎ†¨ÏùÑ ÏòÅÌôîÎ™ÖÏùÑ Îß§ÌïëÌï¥ DataFrameÏúºÎ°ú Î≥ÄÌôò
item_sim_df = pd.DataFrame(data =item_sim, index=ratings_matrix.columns, columns=ratings_matrix.columns)
print(item_sim_df.shape)
item_sim_df.head(3)
```

### ÏïÑÏù¥ÌÖú Í∏∞Î∞ò ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÌòëÏóÖ ÌïÑÌÑ∞ÎßÅÏúºÎ°ú Í∞úÏù∏ÌôîÎêú ÏòÅÌôî Ï∂îÏ≤ú
ÏòÅÌôî Ïú†ÏÇ¨ÎèÑ Îç∞Ïù¥ÌÑ∞Î•º Ïù¥Ïö©Ìï¥ ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÌòëÏóÖ ÌïÑÌÑ∞ÎßÅÏúºÎ°ú Í∞úÏù∏ÏóêÍ≤å ÏµúÏ†ÅÌôîÎêú ÏòÅÌôîÏ∂îÏ≤úÏùÑ Íµ¨ÌòÑ.

Í∞úÏù∏ÌôîÎêú ÏòÅÌôîÏ∂îÏ≤úÏùò Í∞ÄÏû• ÌÅ∞ ÌäπÏßïÏùÄ Í∞úÏù∏Ïù¥ ÏïÑÏßÅ Í¥ÄÎûåÌïòÏßÄ ÏïäÏùÄ ÏòÅÌôîÎ•º Ï∂îÏ≤úÌïúÎã§Îäî Í≤É.

ÏïÑÏßÅ Í¥ÄÎûåÌïòÏßÄ ÏïäÏùÄ ÏòÅÌôîÏóê ÎåÄÌï¥ÏÑú ÏïÑÏù¥ÌÖú Ïú†ÏÇ¨ÎèÑÏôÄ Í∏∞Ï°¥Ïóê Í¥ÄÎûåÌïú ÏòÅÌôîÏùò ÌèâÏ†ê Îç∞Ïù¥ÌÑ∞Î•º Í∏∞Î∞òÏúºÎ°ú Ìï¥ ÏÉàÎ°≠Í≤å Î™®Îì† ÏòÅÌôîÏùò ÏòàÏ∏° ÌèâÏ†êÏùÑ Í≥ÑÏÇ∞Ìïú ÌõÑ ÎÜíÏùÄ ÏòàÏ∏° ÌèâÏ†êÏùÑ Í∞ÄÏßÑ ÏòÅÌôîÎ•º Ï∂îÏ≤úÌïòÎäî Î∞©Ïãù

Í∞úÏù∏ÌôîÎêú ÏòàÏ∏° ÌèâÏ†êÏùÄ Îã§ÏùåÍ≥º Í∞ôÏùÄ ÏãùÏúºÎ°ú Íµ¨Ìï† Ïàò ÏûàÏùå

$\hat{R}_{u,i} = \sum\nolimits_{N} (S_{i,N} \cdot R_{u,N}) \Big/ \sum\nolimits_{N} (|S_{i,N}|)$

* $\hat{R}_{u,i}$: ÏÇ¨Ïö©Ïûê $u$, ÏïÑÏù¥ÌÖú $i$Ïùò Í∞úÏù∏ÌôîÎêú ÏòàÏ∏° ÌèâÏ†ê Í∞í
* $S_{i,N}$: ÏïÑÏù¥ÌÖú$i$ÏôÄ Í∞ÄÏû• Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Top-NÍ∞ú ÏïÑÏù¥ÌÖúÏùò Ïú†ÏÇ¨ÎèÑ Î≤°ÌÑ∞
* $R_{u,N}$: ÏÇ¨Ïö©Ïûê $u$Ïùò ÏïÑÏù¥ÌÖú $i$ÏôÄ Í∞ÄÏû• Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Top‚ÄîNÍ∞ú ÏïÑÏù¥ÌÖúÏóê ÎåÄÌïú Ïã§Ï†ú ÌèâÏ†ê Î≤°ÌÑ∞
  * $N$ Í∞íÏùÄ ÏïÑÏù¥ÌÖúÏùò ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Î≤îÏúÑ Í≥ÑÏàò(item neighbor)Î•º ÏùòÎØ∏



```python
# ÏòÅÌôîÍ∞ÑÏùò Ïú†ÏÇ¨ÎèÑ ÌñâÎ†¨Í≥º ÏÇ¨Ïö©Ïûê-ÏòÅÌôî ÌèâÏ†ê ÌñâÎ†¨ÏùÑ ÌôúÏö©Ìï¥ ÏÇ¨Ïö©ÏûêÎ≥Ñ ÏµúÏ†ÅÌôîÎêú(Í∞úÏù∏ÌôîÎêú) ÌèâÏ†ê Ïä§ÏΩîÏñ¥Î•º ÏòàÏ∏°ÌïòÎäî Ìï®Ïàò ÏÉùÏÑ±
def predict_rating(ratings_arr, item_sim_arr):
    ratings_pred = ratings_arr.dot(item_sim_arr)/np.array([np.abs(item_sim_arr).sum(axis=1)])
    return ratings_pred
```

```python
ratings_pred = predict_rating(ratings_matrix.values, item_sim_df.values)
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index=ratings_matrix.index,
                                   columns=ratings_matrix.columns)
ratings_pred_matrix.head(3)
```

* ÏòàÏ∏° ÌèâÏ†êÏù¥ ÏÇ¨Ïö©ÏûêÎ≥Ñ ÏòÅÌôîÏùò Ïã§Ï†ú ÌèâÏ†êÍ≥º ÏòÅÌôîÏùò ÏΩîÏÇ¨Ïù∏ Ïú†ÏÇ¨ÎèÑÎ•º ÎÇ¥Ï†Å(dot)Ìïú Í∞íÏù¥Í∏∞ ÎïåÎ¨∏Ïóê Í∏∞Ï°¥Ïóê ÏòÅÌôîÎ•º Í¥ÄÎûåÌïòÏßÄ ÏïäÏïÑ 0Ïóê Ìï¥ÎãπÌñàÎçò Ïã§Ï†ú ÏòÅÌôî ÌèâÏ†êÏù¥ ÏòàÏ∏°ÏóêÏÑúÎäî Í∞íÏù¥ Î∂ÄÏó¨ÎêòÎäî Í≤ΩÏö∞Í∞Ä ÎßéÏù¥ Î∞úÏÉù.
* ÎÇ¥Ï†Å Í≤∞Í≥ºÎ•º ÏΩîÏÇ¨Ïù∏ Ïú†ÏÇ¨ÎèÑ Î≤°ÌÑ∞ Ìï©ÏúºÎ°ú ÎÇòÎàÑÏóàÍ∏∞ ÎïåÎ¨∏Ïóê ÏòàÏ∏° ÌèâÏ†êÏù¥ Ïã§Ï†ú ÌèâÏ†êÏóê ÎπÑÌï¥ ÏûëÏùÑ Ïàò ÏûàÏùå.



ÏòàÏ∏° Í≤∞Í≥ºÍ∞Ä ÏõêÎûòÏùò Ïã§Ï†ú ÌèâÏ†êÍ≥º ÏñºÎßàÎÇò Ï∞®Ïù¥Í∞Ä ÏûàÎäîÏßÄ MSEÎ•º ÌÜµÌï¥ ÌååÏïÖ.

Ïã§Ï†úÏôÄ ÏòàÏ∏° ÌèâÏ†êÏùò Ï∞®Ïù¥Îäî Í∏∞Ï°¥Ïóê ÌèâÏ†êÏù¥ Î∂ÄÏó¨Îêú Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ÏÑúÎßå Ïò§Ï∞® Ï†ïÎèÑÎ•º Ï∏°Ï†ï

```python
from sklearn.metrics import mean_squared_error

# ÏÇ¨Ïö©ÏûêÍ∞Ä ÌèâÏ†êÏùÑ Î∂ÄÏó¨Ìïú ÏòÅÌôîÏóê ÎåÄÌï¥ÏÑúÎßå ÏòàÏ∏° ÏÑ±Îä• ÌèâÍ∞Ä MSEÎ•º Íµ¨Ìï®
def get_mse(pred, actual):
    # ÌèâÏ†êÏù¥ ÏûàÎäî Ïã§Ï†ú ÏòÅÌôîÎßå Ï∂îÏ∂ú
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred, actual)

print('ÏïÑÏù¥ÌÖú Í∏∞Î∞ò Î™®Îì† ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ MSE:', get_mse(ratings_pred, ratings_matrix.values))
```

`predict_rating()` Ìï®ÏàòÎäî Î™®Îì† ÏòÅÌôî Í∞ÑÏùò Ïú†ÏÇ¨ÎèÑ Î≤°ÌÑ∞Î•º Ï†ÅÏö©Ìïú Í≤ÉÏù¥Í∏∞ ÎïåÎ¨∏Ïóê ÏÉÅÎåÄÏ†ÅÏúºÎ°ú ÌèâÏ†ê ÏòàÏ∏°Ïù¥ Îñ®Ïñ¥Ïßê.

ÌäπÏ†ï ÏòÅÌôîÏôÄ Í∞ÄÏû• ÎπÑÏä∑Ìïú Ïú†ÏÇ¨ÎèÑÎ•º Í∞ÄÏßÄÎäî ÏòÅÌôîÏóê ÎåÄÌï¥ÏÑúÎßå Ïú†ÏÇ¨ÎèÑ Î≤°ÌÑ∞Î•º Ï†ÅÏö©ÌïòÎäî Ìï®ÏàòÎ°ú Î≥ÄÍ≤Ω `predict_rating_top_sim(ratings_arr, item_sim_arr, n=20)`. Ïù¥Îäî n Ïù∏ÏûêÎ•º Í∞ÄÏßÄÍ≥† ÏûàÏñ¥ÏÑú Top-N Ïú†ÏÇ¨ÎèÑÎ•º Í∞ÄÏßÄÎäî ÏòÅÌôî Ïú†ÏÇ¨ÎèÑ Î≤°ÌÑ∞Îßå ÏòàÏ∏°Í∞íÏùÑ Í≥ÑÏÇ∞ÌïòÎäî Îç∞ Ï†ÅÏö©. ÏàòÌñâÏãúÍ∞ÑÏù¥ Ïò§Îûò Í±∏Î¶∞Îã§Îäî Îã®Ï†êÏù¥ ÏûàÏùå

```python
def predict_rating_topsim(ratings_arr, item_sim_arr, n=20):
    # ÏÇ¨Ïö©Ïûê-ÏïÑÏù¥ÌÖú ÌèâÏ†ê ÌñâÎ†¨ ÌÅ¨Í∏∞ÎßåÌÅº 0ÏúºÎ°ú Ï±ÑÏö¥ ÏòàÏ∏° ÌñâÎ†¨ Ï¥àÍ∏∞Ìôî
    pred = np.zeros(ratings_arr.shape)

    # ÏÇ¨Ïö©Ïûê-ÏïÑÏù¥ÌÖú ÌèâÏ†ê ÌñâÎ†¨Ïùò Ïó¥ ÌÅ¨Í∏∞ÎßåÌÅº Î£®ÌîÑ ÏàòÌñâ
    for col in range(ratings_arr.shape[1]):
        # Ïú†ÏÇ¨ÎèÑ ÌñâÎ†¨ÏóêÏÑú Ïú†ÏÇ¨ÎèÑÍ∞Ä ÌÅ∞ ÏàúÏúºÎ°ú nÍ∞ú Îç∞Ïù¥ÌÑ∞ ÌñâÎ†¨Ïùò Ïù∏Îç±Ïä§ Î∞òÌôò
        top_n_items = [np.argsort(item_sim_arr[:, col])[:-n-1:-1]]
        # Í∞úÏù∏ÌôîÎêú ÏòàÏ∏° ÌèâÏ†êÏùÑ Í≥ÑÏÇ∞
        for row in range(ratings_arr.shape[0]):
            pred[row, col] = item_sim_arr[col, :][top_n_items].dot(ratings_arr[row,:][top_n_items].T)
            pred[row, col] /= np.sum(np.abs(item_sim_arr[col,:][top_n_items]))

    return pred
```

```python
# ÏòàÏ∏° ÌèâÏ†ê Í≥ÑÏÇ∞ ÌõÑ Ïã§Ï†ú ÌèâÏ†êÍ≥ºÏùò MSE Íµ¨ÌïòÍ∏∞
ratings_pred = predict_rating_topsim(ratings_matrix.values, item_sim_df.values, n=20)
print('ÏïÑÏù¥ÌÖú Í∏∞Î∞ò ÏµúÍ∑ºÏ†ë Top-20 Ïù¥ÏõÉ MSE:', get_mse(ratings_pred, ratings_matrix.values))

# Í≥ÑÏÇ∞Îêú ÏòàÏ∏° ÌèâÏ†ê Îç∞Ïù¥ÌÑ∞Îäî DataFrameÏúºÎ°ú Ïû¨ÏÉùÏÑ±
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index=ratings_matrix.index,
                                   columns=ratings_matrix.columns)
```

```python
# ÏÇ¨Ïö©ÏûêÍ∞Ä ÌèâÏ†êÏùÑ Ï£ºÏßÄ ÏïäÏùÄ ÏòÅÌôîÎ•º Î¶¨Ïä§Ìä∏Î°ú Î∞òÌôòÌïòÎäî Ìï®Ïàò ÏÉùÏÑ±
def get_unseen_movies(ratings_matrix, userId):
    user_rating = ratings_matrix.loc[userId, :]
    # ÏÇ¨Ïö©ÏûêÍ∞Ä Ïù¥ÎØ∏ Í¥ÄÎûåÌïú ÏòÅÌôî Î¶¨Ïä§Ìä∏
    already_seen = user_rating[user_rating > 0].index.tolist()
    # Î™®Îì† ÏòÅÌôîÎ™ÖÏùÑ Î¶¨Ïä§Ìä∏ Í∞ùÏ≤¥Î°ú ÏÉùÏÑ±
    movies_list = ratings_matrix.columns.tolist()
    # list comprehensionÏúºÎ°ú ÏïÑÏßÅ Í¥ÄÎûåÌïòÏßÄ ÏïäÏùÄ ÏòÅÌôî Î¶¨Ïä§Ìä∏ ÏÉùÏÑ±
    unseen_list = [ movie for movie in movies_list if movie not in already_seen]

    return unseen_list
```

```python
def recomm_movie_by_user_id(pred_df, userId, unseen_list, top_n=10):
    # ÏòàÏ∏° ÌèâÏ†ê DataFrameÏóêÏÑú ÏÇ¨Ïö©ÏûêidÏôÄ unseen listÎ°ú Îì§Ïñ¥Ïò® ÏòÅÌôîÎ™Ö ÏπºÎüºÏùÑ Ï∂îÏ∂úÌï¥
    # Í∞ÄÏû• ÏòàÏ∏° ÌèâÏ†êÏù¥ ÎÜíÏùÄ ÏàúÏúºÎ°ú Ï†ïÎ†¨
    recomm_movies = pred_df.loc[userId, unseen_list].sort_values(ascending=False)[:top_n]
    return recomm_movies

# ÏÇ¨Ïö©ÏûêÍ∞Ä Í¥ÄÎûåÌïòÏßÄ ÏïäÏùÄ ÏòÅÌôîÎ™Ö Ï∂îÏ∂ú
unseen_list = get_unseen_movies(ratings_matrix, 9)

# ÏïÑÏù¥ÌÖú Í∏∞Î∞òÏùò ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÌòëÏóÖ ÌïÑÌÑ∞ÎßÅÏúºÎ°ú ÏòÅÌôî Ï∂îÏ≤ú
recomm_movies = recomm_movie_by_user_id(ratings_pred_matrix, 9, unseen_list, top_n=10)

# ÌèâÏ†ê Îç∞Ïù¥ÌÑ∞Î•º DFÎ°ú ÏÉùÏÑ±
recomm_movies = pd.DataFrame(data=recomm_movies.values, index=recomm_movies.index,
                             columns=['pred_score'])

recomm_movies
```
